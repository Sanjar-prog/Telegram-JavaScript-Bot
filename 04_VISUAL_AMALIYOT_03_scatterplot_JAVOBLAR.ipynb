## 3-AMALIYOT. `scatterplot`


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Ushbu amaliyotda uybor.uz saytidan yuklab olingan ma'lumotlar tahlili bilan shu'gullanamiz.

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/blob/main/uybor_scrapping.csv?raw=true")
df.head()
```

#### Yuqoridagi jadvalda Toshkent shahridagi sotildagian uylar haqida ma'lumotlar jamlangan.

### 1-VAZIFA. Uylarning narxi (`price`) va maydoni (`size`) o'rtasida bog'liqlikni ko'rsatuvchi scatter plot chizing.

```python
#JAVOBINGIZNI SHUYERGA YOZING
plt.figure(figsize=(10, 6))
plt.scatter(df['size'], df['price'], alpha=0.5)
plt.title('Uylarning maydoni va narxi o\'rtasidagi bog\'liqlik')
plt.xlabel('Uy maydoni (kv.m)')
plt.ylabel('Narx (USD)')
plt.grid(True)
plt.show()
```

```python
#@title Kutilgan natija
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='size', y='price')
plt.show()
```

### Jadvalda 7000 dan oshiq uy bor, aksar uylarning maydoni 250kv.m dan kam. Keling noodatiy katta qiymatlarni tashlab yuboramiz va yuqoridagi chizmani qaytadan chizamiz.

```python
#JAVOBINGIZNI SHUYERGA YOZING
df_filtered = df[df['size'] <= 250]

# Filterlangan jadvalni ko'rib chiqamiz
print(df_filtered.head())

# Scatter plot chizamiz
plt.figure(figsize=(10, 6))
plt.scatter(df_filtered['size'], df_filtered['price'], alpha=0.5)
plt.title('250 kv.m dan kichik uylarning maydoni va narxi o\'rtasidagi bog\'liqlik')
plt.xlabel('Uy maydoni (kv.m)')
plt.ylabel('Narx (USD)')
plt.grid(True)
plt.show()
```

```python
#@title Kutilgan natija
df = df[df['size']<250]
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='size', y='price')
plt.show()
```

### 2-VAZIFA. Uylarning narxi (price) va maydoni (size) o'rtasida bog'liqlikni ko'rsatuvchi scatter va to'g'ri chiziqni chizing.

```python
# JAVOB UCHUN JOY
plt.figure(figsize=(10, 6))
sns.regplot(x='size', y='price', data=df_filtered, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('250 kv.m dan kichik uylar uchun narx va maydon o\'rtasidagi bog\'liqlik')
plt.xlabel('Uy maydoni (kv.m)')
plt.ylabel('Narx (USD)')
plt.grid(True)
plt.show()
```

```python
#@title Kutilgan natija
plt.figure(figsize=(10,6))
sns.regplot(data=df, x='size', y='price')
plt.show()
```

### 3-VAZIFA. Uylarning narxi, maydoni va xonalar soni o'rtasidagi bog'liqlikni ko'rsating.

```python
#@title Kutilgan natija
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='size', y='price', hue='rooms')
plt.show()
```

### 4-VAZIFA. Uylarning narxi, maydoni va tumanlar o'rtasidagi bog'liqlikni ko'rsating.

```python
# JAVOBNI SHUYERGA YOZING
plt.figure(figsize=(14, 8))
sns.scatterplot(x='size', y='price', hue='district', data=df_filtered, palette='tab10', alpha=0.7)
plt.title('Narx, maydon va tumanlar o\'rtasidagi bog\'liqlik')
plt.xlabel('Uy maydoni (kv.m)')
plt.ylabel('Narx (USD)')
plt.legend(title='Tuman', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
#@title Kutilgan natija
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='size', y='price', hue='district')
plt.show()
```

### 5-VAZIFA. Swarm ploit yordamida turli tumanlardag uylarning narxini taqqoslang.

```python
# JAVOB UCHUN JOY
plt.figure(figsize=(14, 8))
sns.swarmplot(x='district', y='price', data=df_filtered, size=3)
plt.title('Tumanlar bo\'yicha uy narxlarini taqqoslash (Swarm plot)')
plt.xlabel('Tuman')
plt.ylabel('Narx (USD)')
plt.xticks(rotation=45)  # Tuman nomlarini qiyshiq chiqaramiz, o'qish osonroq bo'ladi
plt.grid(True)
plt.show()
```

```python
#@title Kutilgan natija
plt.figure(figsize=(10,6))
sns.swarmplot(data=df, x='district', y='price')
plt.xticks(rotation=90)
plt.show()
```
