```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Ushbu amaliyotda uybor.uz saytidan yuklab olingan ma'lumotlar tahlili bilan shu'gullanamiz.

```python
df = pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/blob/main/uybor_scrapping.csv?raw=true")
df.head()
```

#### Yuqoridagi jadvalda Toshkent shahridagi sotildagian uylar haqida ma'lumotlar jamlangan.

## 1-VAZIFA. Jadvaldan xonalari 5 xonadan ko'p bo'lgan, narxi 200000 mingdan, maydoni 250kv.m dan ortiq bo'lgan uylarni tashlab yuboring

```python
#JAVOBINGIZNI SHUYERGA YOZING
df = df[(df['rooms'] <= 5) & (df['price'] <= 200000) & (df['size'] <= 250)]
df.head()
```

```python
#@title Natija
df=df[df['rooms']<6]
df=df[df['size']<250]
df=df[df['price']<200000]
df.shape
```

## 2-VAZIFA. `subplot` yordamida uylarning narxi va xonalar soni taqsimoti grafiklarini yonma yon chizing

```python
#JAVOBINGIZNI SHUYERGA YOZING
plt.figure(figsize=(12, 5))

# 1-grafik: Uy narxlari taqsimoti
plt.subplot(1, 2, 1)
plt.hist(df['price'], bins=30, color='skyblue', edgecolor='black')
plt.title('Uy narxlari taqsimoti')
plt.xlabel('Narx')
plt.ylabel('Uy soni')

# 2-grafik: Xonalar soni taqsimoti
plt.subplot(1, 2, 2)
plt.hist(df['rooms'], bins=10, color='salmon', edgecolor='black')
plt.title('Xonalar soni taqsimoti')
plt.xlabel('Xonalar soni')
plt.ylabel('Uy soni')

plt.tight_layout()
plt.show()
```

```python
#@title Kutilgan natija
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.histplot(data=df, x='price')
plt.subplot(1,2,2)
sns.histplot(data=df, x='size')
plt.show()
```

### 2-VAZIFA. Uylarning narxi taqsimoti va narx va maydoni o'rtasidagi bog'liqlik grafiklarini bitta chizmada ustma ust chiqaring.

```python
# JAVOB UCHUN JOY
plt.figure(figsize=(10, 8))

# 1-chi grafik: Narx taqsimoti (histogramma)
plt.subplot(2, 1, 1)
plt.hist(df['price'], bins=30, color='skyblue', edgecolor='black')
plt.title('Uy narxlari taqsimoti')
plt.xlabel('Narx')
plt.ylabel('Uy soni')

# 2-chi grafik: Narx va maydon o‘rtasidagi bog‘liqlik (scatter)
plt.subplot(2, 1, 2)
plt.scatter(df['size'], df['price'], alpha=0.5, color='green')
plt.title('Maydon va narx o‘rtasidagi bog‘liqlik')
plt.xlabel('Maydon (kv.m)')
plt.ylabel('Narx')

plt.tight_layout()
plt.show()
```

```python
#@title Kutilgan natija
fig, ax = plt.subplots(2, 1, figsize=(10,10))
# Umumiy chizma nomini beramiz:
sns.histplot(ax=ax[0], data=df, x='price')
sns.scatterplot(ax=ax[1], data=df, x=df['size'], y=df['price'])
plt.show()
```

### 3-VAZIFA. Quyidagi grafikni chizing. Ahamiyat bering, oxirgi grafikda tumanlar bo'yicha uylarning o'rtacha narxi tasvirlangan.

```python
# JAVOB UCHUN JOY
# Tumanlar bo‘yicha uylarning o‘rtacha narxini hisoblaymiz
district_avg_price = df.groupby('district')['price'].mean().sort_values(ascending=False)

# Grafikni chizamiz
plt.figure(figsize=(12, 6))
sns.barplot(x=district_avg_price.values, y=district_avg_price.index, palette='viridis')
plt.title("Tumanlar bo‘yicha uylarning o‘rtacha narxi")
plt.xlabel("O‘rtacha narx")
plt.ylabel("Tuman")

plt.tight_layout()
plt.show()
```

```python
#@title Kutilgan natija

# Tumanlar bo‘yicha uylarning o‘rtacha narxini hisoblaymiz
district_avg_price = df.groupby('district')['price'].mean().sort_values(ascending=False)

# Grafikni chizamiz
plt.figure(figsize=(12, 6))
sns.barplot(x=district_avg_price.values, y=district_avg_price.index, palette='viridis')
plt.title("Tumanlar bo‘yicha uylarning o‘rtacha narxi")
plt.xlabel("O‘rtacha narx")
plt.ylabel("Tuman")

plt.tight_layout()
plt.show()
```
