```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns # agar seaborn o'rnatilmagan bo'lsa pip install seaborn
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv')
df.head()
```

```python
df = df[df.year>2000]
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='year', y='price')
sns.regplot(data=df, x='year', y='price')
plt.show()
```

```python
df = df[df.year>2000]
plt.figure(figsize=(16,10))
sns.scatterplot(data=df, x='mileage', y='price', hue='fuelType', s=100)
# sns.regplot(data=df, x='mileage', y='price')
plt.show()
```

```python
sns.swarmplot(data=df, x=df['fuelType'], y=df['price'])
plt.show()
```
