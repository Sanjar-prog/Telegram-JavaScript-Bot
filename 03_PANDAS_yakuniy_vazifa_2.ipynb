```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/Electricity_Production_By_Source.csv")
df.columns = ['davlat','kod','yil','komir','gaz','gidro','yangilanuvchi','quyosh','neft','shamol','atom']
df.head()
```

```python
# JAVOBNI SHUYERGA YOZING
# DF hajmini chiqarish (qator va ustun soni)
print(df.shape)
```

```python
# JAVOBNI SHUYERGA YOZING
# DF dagi barcha unikal davlatlarni chiqarish
davlatlar = df['davlat'].unique()
print(davlatlar)
```

### 3-VAZIFA. DF dan faqat AQSh (`United States`) uchun ma'lumotlarni ajratib oling va yangi `dfus` nomli o'zgaruvchiga saqlang

```python
# JAVOBNI SHUYERGA YOZING
# United States ma'lumotlarini ajratib olish
dfus = df[df['davlat'] == 'United States']

# Natijani ko'rish
print(dfus.head())  # Dastlabki 5 qatorni chiqarish
```

```python
#@title Kutilgan natija
dfus.head(6)
```

### 4-VAZIFA. `dfus` jadvalining indeksini `yil` ustuni bilan o'zgartiring. DF dan `yil` ustunini tashlab yuboring.

```python
# JAVOBNI SHUYERGA YOZING# Yil ustunini indeks sifatida belgilash
dfus = dfus.set_index('yil')

# Yil ustunini tashlab yuborish (allaqachon indeks bo'lganligi sababli bu bosqich shart emas)
# dfus = dfus.drop(columns=['yil'])  # Agar indeks qilib belgilamasdan o'chirish kerak bo'lsa

# Natijani ko'rish
print(dfus.head())  # Dastlabki 5 qatorni chiqarish
```

```python
#@title Kutilgan natija
```

### 5-VAZIFA. AQSh 2020-yilda jami qancha elektr energiyasi ishlab chiqqan?

```python
# JAVOBNI SHUYERGA YOZING
# JAVOBNI SHUYERGA YOZING
# 2020-yil bo'yicha ma'lumotni ajratib olish
aqsh_2020 = dfus.loc[2020]  # Yil indeks sifatida belgilangan

# Jami elektr energiyasini hisoblash (faqat raqamli ustunlar yig‘indisi)
jami_energiyasi = aqsh_2020[['komir', 'gaz', 'gidro', 'yangilanuvchi', 'quyosh', 'neft', 'shamol', 'atom']].sum().sum()
#jami_energiyasi = aqsh_2020.drop(columns=['davlat', 'kod']).sum().sum() # Alternatively, exclude 'davlat' and 'kod' columns

# Natijani chiqarish
print(f"AQSh 2020-yilda jami {jami_energiyasi:.2f} teravatt-soat elektr energiya ishlab chiqargan.")
```

### 6-VAZIFA. 2010-2020-yillarda AQSH turli manbaalardan o'rtacha qancha elektr toki olganini chiqaring. Natijani kichikdan kattaga qarab tartiblang.

```python
# JAVOBNI SHUYERGA YOZING
# 2010-2020 yillar oralig'idagi ma'lumotlarni ajratib olish
dfus_2010_2020 = dfus.loc[2010:2020]

# Har bir manbadan o'rtacha ishlab chiqarilgan elektr energiyasini hisoblash
# Only include numeric columns for calculating the mean
ortalama_energiyasi = dfus_2010_2020[['komir', 'gaz', 'gidro', 'yangilanuvchi', 'quyosh', 'neft', 'shamol', 'atom']].mean()

# Kichikdan kattaga qarab tartiblash
ortalama_energiyasi = ortalama_energiyasi.sort_values()

# Natijani chiqarish
print(ortalama_energiyasi)
```

### 7-VAZIFA. AQSH gidro elektr stansiyalardan eng kam va eng ko'p elektr energiyasi olgan yillarni toping.

```python
# JAVOBNI SHUYERGA YOZING
# Gidroelektr energiyasi ustunidagi eng kichik va eng katta qiymatlarni topish
min_yil = dfus['gidro'].idxmin()  # Eng kam ishlab chiqarilgan yil
max_yil = dfus['gidro'].idxmax()  # Eng ko'p ishlab chiqarilgan yil

# Eng kam va eng ko‘p gidro energiya ishlab chiqarilgan qiymatlar
min_qiymat = dfus.loc[min_yil, 'gidro']
max_qiymat = dfus.loc[max_yil, 'gidro']

# Natijani chiqarish
print(f"AQSh gidro elektr stansiyalaridan eng kam {min_qiymat:.2f} TWh elektr energiya olgan yil: {min_yil}")
print(f"AQSh gidro elektr stansiyalaridan eng ko'p {max_qiymat:.2f} TWh elektr energiya olgan yil: {max_yil}")
```

### 8-VAZIFA.

### 10-VAZIFA.
#### 10.1. Asl DF dan 2000-2020-yillar, `France` uchun `atom`, `komir`, `gaz`, `quyosh` manbaalarini alohida `dffr` jadvaliga ko'chirib oling
#### 10.2. `dffr` jadvalini yil ustuni bo'yicha tartiblang

```python
# JAVOBNI SHUYERGA YOZING
# 2000-2020-yillar oralig'ida France davlatiga tegishli ma'lumotlarni ajratib olish
dffr = df[(df['davlat'] == 'France') & (df['yil'].between(2000, 2020))].copy()

# Faqat kerakli ustunlarni tanlash
dffr = dffr[['yil', 'atom', 'komir', 'gaz', 'quyosh']]

# Natijani ko‘rish
print(dffr.head())
```

```python
#@title Kutilgan natija
```

#### 10.3 `dffr`uchun Korrelyasiya jadvalini chiqaring.

```python
# JAVOBNI SHUYERGA YOZING
# dffr jadvali uchun korrelyatsiya matritsasi
correlation_matrix = dffr.corr()

# Natijani chiqarish
print(correlation_matrix)
```
